<h1> Exercice </h1>

Avec le DataFrame events suivant qui enregistre des dates d'événements liés à des sorties de films, effectuez les opérations suivantes :
1) Convertissez les colonnes 'Release_Date' et 'Production_Date' de texte en dates.
2) Extrayez l'année, le mois et le jour des dates de sortie des films.
3) Déterminez la semaine de l'année et le jour de la semaine pour la date de sortie de chaque film.
4) Calculez la différence en jours entre la 'Release_Date' et la 'Production_Date' pour chaque film.
5) Pour chaque film, calculez la différence en jours depuis l'événement précédent dans le DataFrame.
6) Ajoutez 2 semaines à la 'Release_Date' de chaque film pour estimer la date de sortie en vidéo.
7) Filtrez pour obtenir uniquement les films dont la date de sortie est après le 2023-02-01.
8) Identifier les films dont la 'Release_Date' tombe un samedi ou un dimanche
9) Sélectionner les films produits avant le 01-01-2023 mais sortis après cette date.

```python
import pandas as pd
import numpy as np

movies = pd.read_csv('movies_light.csv')
movies.head(4)
```

```python
# 1. Conversion des colonnes de texte en dates
movies['Release_Date'] = pd.to_datetime(movies['Release_Date'])
movies['Production_Date'] = pd.to_datetime(movies['Production_Date'])
movies.head()
```

```python
# 2. Extraction de l'année, du mois et du jour
movies['Release_Year'] = movies['Release_Date'].dt.year
movies['Release_Month'] = movies['Release_Date'].dt.month
movies['Release_Day'] = movies['Release_Date'].dt.day
movies.head()
```

```python
# 3. Semaine de l'année et jour de la semaine
movies['Release_Week'] = movies['Release_Date'].dt.isocalendar().week
movies['Release_Weekday'] = movies['Release_Date'].dt.day_name()
movies.head()
```

```python
# 4. Différence de temps en jours
movies['Time_Difference'] = (movies['Release_Date'] - movies['Production_Date']).dt.days
movies.head()
```

```python
# 5. Différence en jours depuis l'événement précédent
movies['Time_Difference_Previous'] = movies['Time_Difference'].diff()
movies.head()
```

```python
# 6. Ajout de temps (2 semaines) à la colonne de date
movies['Release_Date_Extended'] = movies['Release_Date'] + pd.DateOffset(weeks=2)
movies.head()
```

```python
# 7. Filtre sur la colonne de dates
movies_filtered = movies[movies['Release_Date'] > '2023-02-01']
movies_filtered
```

```python
#8. Films dont la date de sortie est un week-end
movies_weekend = movies[movies['Release_Weekday'].isin(['Saturday', 'Sunday'])]
movies_weekend
```

```python
#9 Sélectionner les films produits avant le 01-01-2023 mais sortis après cette date.
movies_filtered = movies[(movies['Production_Date'] < '2023-01-01') & (movies['Release_Date'] > '2023-01-01')]
movies_filtered
```
