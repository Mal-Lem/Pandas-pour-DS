Vous travaillez pour une société de distribution de films. Vous avez un DataFrame movies qui contient des données sur les sorties de films les plus rentables. Vous devez filtrer, trier et transformer ces données pour préparer votre analyse.

1) Calculez le total des recettes (Total Gross) de tous les films.
2) Déterminez le revenu moyen pat théatre (Average) et le revenu médian pour tous les films.
3) Trouvez les recettes d'ouverture maximales (Opening) et minimales.
4) Comptez le nombre de fois où chaque Distributor apparaît dans le dataset.
5) Identifiez le nombre de valeurs uniques dans la colonne Genre.
6) Obtenez la distribution du pourcentage des recettes totales (% of Total) avec des quartiles.
7) Calculez l'écart-type (std) et la kurtosis des cinémas (theaters).
8) Groupez les données par Distributor et calculez le chiffre d'affaires moyen.
9) Utilisez cut() pour catégoriser les films en fonction de leurs revenus par salle en 'Low', 'Moderate', 'High', et 'Very High' en fonction des quartiles de distribution de cette variable.
10) Créez une pivot table qui montre le chiffre d'affaires moyen par Distributor et par revenu par salle (colonne créée à la questions précédente).
11) Binarisez la colonne Total Gross en deux catégories - au-dessus et en dessous de la médiane - avec cut().
12) Calculez la covariance et la corrélation entre Total Gross et Opening.
13) Calculez la covariance et la corrélation entre Total Gross et Theaters.

```python
import pandas as pd

movies = pd.read_csv('Top_Highest_Openings_Modeled.csv')
movies.head(3)
```

```python
# 1. Total des recettes
movies['Total Gross'].sum()
```

```python
# 2. Salaire moyen et médian
movies['Average'].mean()
```

```python
# 3. Recettes d'ouverture maximales et minimales
movies['Opening'].max()
movies['Opening'].min()
```

```python
# 4. Nombre d'apparitions par Distributeur
movies['Distributor'].value_counts()
```

```python
# 5. Nombre de valeurs uniques dans Genre
movies['Genre'].nunique()
```

```python
# 6. Distribution de % of Total avec quartiles
movies['% of Total'].quantile([0.25, 0.5, 0.75])
```

```python
# 7. Écart-type et kurtosis des cinémas
movies['Theaters'].std()
movies['Theaters'].kurtosis()
```

```python
# 8. Chiffre d'affaires moyen par Distributeur
movies.groupby('Distributor')['Total Gross'].mean()
```

```python
# 9: Catégorisation des revenus par salle
movies['Revenus par salle '] = pd.cut(movies['Average'],bins=4, labels=['Low', 'Moderate', 'High', 'Very High'])
movies.head()
```

Créez une pivot table qui montre le chiffre d'affaires moyen par Distributor et par revenu par salle (colonne créée à la questions précédente). 11) Binarisez la colonne Total Gross en deux catégories - au-dessus et en dessous de la médiane - avec cut().

```python
# 10. Pivot table avec chiffre d'affaires moyen par Distributeur et Genre
pivot_table = pd.pivot_table(movies, values='Total Gross', index='Distributor', columns='Genre', aggfunc='mean')
pivot_table
```

```python
# 11. Binarisation de Total Gross avec cut
movies['Total Gross Binary'] = pd.cut(movies['Total Gross'], bins=[0, movies['Total Gross'].median(), movies['Total Gross'].max()], labels=['Low', 'High'])
movies.head()
```

12) Calculez la covariance et la corrélation entre Total Gross et Opening. 13) Calculez la covariance et la corrélation entre Total Gross et Theaters.

```python
# 12. Corrélation entre Total Gross et Opening
movies['Total Gross'].corr(movies['Opening'])
```

```python
# 13. Covariance entre Total Gross et Theaters
movies['Total Gross'].corr(movies['Theaters'])
```
