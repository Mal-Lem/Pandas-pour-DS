<h1> Exercice </h1>

Vous avez deux DataFrames, df_films et df_revenue. Le DataFrame df_films contient des identifiants de films, les titres et l'année de sortie. Le DataFrame df_revenue contient les identifiants de films avec les recettes totales et les recettes domestiques.

1) Mergez df_films et df_revenue afin de ne conserver que les films présents dans les 2 data sets
2) Mergez df_films et df_revenue pour conserver tous les films, même ceux sans revenus.
3) Mergez df_films et df_revenue pour conserver tous les enregistrements de revenus, même ceux sans film correspondant.
4) Mergez df_films et df_revenue en utilisant un pour voir tous les films et tous les revenus, même sans correspondance

```python
import pandas as pd

# Premier DataFrame : informations sur les films
df_films = pd.DataFrame({
    'film_id': [1, 2, 3, 4],
    'title': ['Avengers: Endgame', 'Avatar', 'Titanic', 'Star Wars: The Force Awakens'],
    'release_year': [2019, 2009, 1997, 2015]
})

# Second DataFrame : recettes des films
df_revenue = pd.DataFrame({
    'film_id': [2, 3, 4, 5],
    'total_revenue': [2790, 2187, 2068, 1500],  # En millions pour l'exemple
    'domestic_revenue': [760, 659, 936, 500]  # En millions pour l'exemple
})
```

```python
# 1.
df_merged = pd.merge(df_films,df_revenue,on= 'film_id',how='inner')
df_merged
```

```python
# 2.
df_merged = pd.merge(df_films,df_revenue,on= 'film_id',how='outer')
df_merged
```

```python
# 3.
df_merged = pd.merge(df_films,df_revenue,on= 'film_id',how='right')
df_merged
```

```python
# 4.
df_merged = pd.merge(df_films,df_revenue,on= 'film_id',how='outer',indicator=True)
df_merged
```
